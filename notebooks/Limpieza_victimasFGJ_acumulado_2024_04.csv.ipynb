```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\rubig\\Downloads\\victimasFGJ_acumulado_2024_04.csv")
df
```

```python
# Ver nombres de las columnas para eliminar las no necesarias
df.columns
```

```python
# Eliminar columnas 
df.drop(columns=['anio_inicio', 'mes_inicio', 'fecha_inicio', 'hora_inicio', 'anio_hecho', 'mes_hecho',
                 'hora_hecho', 'sexo', 'edad', 'tipo_persona', 'calidad_juridica', 'competencia', 'colonia_hecho',
                 'colonia_catalogo', 'alcaldia_hecho', 'alcaldia_catalogo', 'municipio_hecho', 'categoria_delito'], inplace=True)
```

```python
df
```

```python
# Palabra clave para después filtrar por filas los delitos que son robo a negocio
palabra_clave = 'NEGOCIO'
```

```python
df = df[df['delito'].str.contains(palabra_clave, case=False, na=False)]
df
```

```python
# Delitos que contienen "robo a negocio"
df['delito'].unique().tolist()
```

```python
#Renombrar columnas
df.rename(columns={'latitud':'y', 'longitud':'x', 'fecha_hecho':'date_single'}, inplace=True)
df
```

```python
#Descartamos columna de delitos y reordenamos
df = df[['x', 'y', 'date_single']]
df
```
